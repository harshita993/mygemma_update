```python
import undetected_chromedriver as uc
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import time

driver = uc.Chrome()
driver.maximize_window()

def scroll_to_bottom(driver):
    # Scroll down to the bottom of the page
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(2)  # Adjust the sleep time as needed

def get_artist_details(artwork_url):
    try:
        driver.get(artwork_url)
        time.sleep(2)
        
        WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CLASS_NAME, 'userName--owby3')))
        soup = BeautifulSoup(driver.page_source, 'html.parser')
        
        # Extract artist name and construct artist URL
        artist_name_element = soup.find('a', class_='userName--owby3')
        if artist_name_element:
            artist_name_href = artist_name_element.get('href')
            artist_url = 'https://pixabay.com' + artist_name_href
            print(f'Artist URL: {artist_url}')
            
            # Navigate to artist's page
            driver.get(artist_url)
            time.sleep(2)
            
            # Scrape artist details
            soup = BeautifulSoup(driver.page_source, 'html.parser')
            artist_name = soup.find('h1', class_='username--CC+CN').text.strip() if soup.find('h1', class_='username--CC+CN') else 'Not Found'
            artist_bio = soup.find('div', class_='metaData--cokvg').text.strip() if soup.find('div', class_='metaData--cokvg') else 'Not Found'
            artist_follow_status = soup.find('div', class_='followStats--gPL7e').text.strip() if soup.find('div', class_='followStats--gPL7e') else 'Not Found'
            
            print(f'Artist: {artist_name}')
            print(f'Bio: {artist_bio}')
            print(f'Follow Status: {artist_follow_status}')
            print(" ")
            time.sleep(3)
            
    except Exception as e:
        
        pass

def main(url):
    try:
        driver.get(url)
        WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CLASS_NAME, 'link--WHWzm')))
        soup = BeautifulSoup(driver.page_source, 'html.parser')
        
        page_count = 0
        while True:
            scroll_to_bottom(driver)  # Scroll to load more content
            time.sleep(2)
            page_count += 1
            print(f'Scrolling down to load more content. Page {page_count}')
            
            # Extract artwork links
            artwork_links = [a['href'] for a in soup.find_all('a', class_='link--WHWzm')]
            
            for link in artwork_links:
                artwork_url = f'https://pixabay.com{link}'
                print(f'Fetching details for artwork: {artwork_url}')
                get_artist_details(artwork_url)
                time.sleep(2)  
            
            # Check for next page button
            next_button = driver.find_element(By.XPATH, "//a[@title='Next']")
            if next_button and not next_button.get_attribute('aria-disabled'):
                next_button.click()
                time.sleep(3)  # Adjust sleep time to allow next page to load
                soup = BeautifulSoup(driver.page_source, 'html.parser')
            else:
                print("No more next pages found. Exiting...")
                break
    
    except Exception as e:
        
        pass

url = 'https://pixabay.com/photos/search/?order=ec'    
main(url)
```
