```python
import requests
from bs4 import BeautifulSoup
from selenium.webdriver.common.by import By
import undetected_chromedriver as uc
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import pandas as pd
from selenium.webdriver.common.keys import Keys
driver = uc.Chrome()
driver.maximize_window()

def get_artist_details(artwork_url):
    try:
        driver.get(artwork_url)
        time.sleep(2)
        WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CLASS_NAME, 'userName--owby3')))
        
        soup = BeautifulSoup(driver.page_source, 'html.parser')
        artist_name = soup.find('a',class_='userName--owby3').get('href')
        artist_url = 'https://pixabay.com'+artist_name
        
       
        driver.get(artist_url)       
        soup = BeautifulSoup(driver.page_source, 'html.parser')
        artist_name_element = soup.find('h1', class_='username--CC+CN')
        artist_bio_element = soup.find('div', class_='metaData--cokvg')
        artist_follow_status_element = soup.find('div', class_='followStats--gPL7e')
                
                
        artist_name = artist_name_element.text.strip() if artist_name_element else 'Not Found'
        artist_bio = artist_bio_element.text.strip() if artist_bio_element else 'Not Found'
        artist_follow_status = artist_follow_status_element.text.strip() if artist_follow_status_element else 'Not Found'
        print(f'Artist: {artist_name}')
        print(f'Bio: {artist_bio}')
        print(f'Follow Status: {artist_follow_status}')
        print(" ")
        time.sleep(3) 
        
    except:
        pass

def page_process(url):
    driver.get(url)    
    driver.implicitly_wait(5)
    try:
        WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CLASS_NAME, 'link--WHWzm')))
        soup = BeautifulSoup(driver.page_source, 'html.parser')
        
       
        artwork_links = [a['href'] for a in soup.find_all('a', class_='link--WHWzm')]
        
        for link in artwork_links:
            artwork_url = f'https://pixabay.com{link}'
            print(f'Fetching details for artwork: {artwork_url}')
            get_artist_details(artwork_url)
            time.sleep(2)  
            
       
        next_pages = [a['href'] for a in soup.find_all('a', class_='buttonBase--r4opq secondaryButton--xk9cO base--jzyee light--uBcBI')]
       
        for next_page in next_pages:
            next_page_url = 'https://pixabay.com' + next_page
            print(f'Navigating to next page: {next_page_url}')
            page_process(next_page_url)
            time.sleep(2) 
            
            
        
    except:
           pass 
def main():
    url = 'https://pixabay.com/photos/search/?order=ec'
    driver.get(url)
    try:
        page_process(url)
    except:
        pass
main()
```
