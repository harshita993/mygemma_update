```python
import requests
from bs4 import BeautifulSoup
from selenium.webdriver.common.by import By
import undetected_chromedriver as uc
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import pandas as pd
from selenium.webdriver.common.keys import Keys
driver = uc.Chrome()
driver.maximize_window()


def fetch_artist_urls(sitemap_url):
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/115.0.0.0 Safari/537.36"
    }
    response = requests.get(sitemap_url, headers=headers)
    soup = BeautifulSoup(response.content, 'xml')
    artist_locs = soup.select('loc:contains("/entity/")')
    return [loc.text for loc in artist_locs]

def get_artist_info(artist_url):
    driver.get(artist_url)
    wait = WebDriverWait(driver, 10)
    soup_artist = BeautifulSoup(driver.page_source, 'html.parser')
    artist_name = soup_artist.find('h1').text if soup_artist.find('h1') else "Unknown"
    print(f"Artist Name: {artist_name}")
    
    date = soup_artist.find('h2').text if soup_artist.find('h2') else "N/A"
    print(f"Date: {date}")
    
    return soup_artist

def get_asset_info(asset_url):
    print("_____________________________________________________")
    driver.get(asset_url)
    wait = WebDriverWait(driver, 10)
    wait.until(EC.presence_of_element_located((By.CLASS_NAME, 'XD0Pkb')))
    soup_asset = BeautifulSoup(driver.page_source, 'html.parser')
    

    metadata = {}
    for li in soup_asset.find_all('li', class_='XD0Pkb'):
        span = li.find('span', class_='PUhAff')
        if span:
            key = span.text.replace(':', '').strip()
            value = li.get_text(separator=' ').replace(span.text, '').strip()
            if 'Link' in key:
                a_tag = li.find('a')
                if a_tag:
                    value = a_tag.get('href')
            metadata[key] = value
    
    for key, value in metadata.items():
        print(f"{key}: {value}")

def main():
    
    
    for i in range(224):
        formatted_number = str(i).zfill(3)
        sitemap_url = f'https://www.gstatic.com/culturalinstitute/sitemaps/artsandculture_google_com/sitemap-{formatted_number}.xml'
        print(f"Processing sitemap: {formatted_number}")
        
        artist_urls = fetch_artist_urls(sitemap_url)
        
        for artist_url in artist_urls:
            print("****************************************************")
            print(f"Artist URL: {artist_url}")
            
            try:
                soup_artist = get_artist_info(artist_url)
                asset_urls = soup_artist.find_all('a', class_='e0WtYb kdYEFe ZEnmnd lXkFp PJLMUc')
                for asset in asset_urls:
                    asset_url = 'https://artsandculture.google.com' + asset.get('href')
                    print(f"asset : {asset_url}")
                    get_asset_info(asset_url)
            except:
                print(f"Error processing {artist_url}")
    
main()
```
