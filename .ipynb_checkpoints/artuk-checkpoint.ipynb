```python
import requests
from bs4 import BeautifulSoup
from selenium.webdriver.common.by import By
import undetected_chromedriver as uc
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import pandas as pd
from selenium.webdriver.common.keys import Keys

driver = uc.Chrome()

driver.maximize_window()

```

```python
def scrap_artist(artist_url):
    driver.get(artist_url)
    wait = WebDriverWait(driver, 10)  # Adjusted wait time for consistency
    soup_art = BeautifulSoup(driver.page_source, 'html.parser')
    
    header = soup_art.find('div', class_='page-header')
    if header:
        artist_name = header.find('h1', class_='page-header-title')
        if artist_name:
            print("Artist Name:", artist_name.text.strip())
        
        p_tags = header.find_all('p')
        if len(p_tags) > 1:
            print("Birth year:", p_tags[1].text.strip())
        if len(p_tags) > 2:
            print("Nationality:", p_tags[2].text.strip())
    
    def scrap_artworks():
        soup_art = BeautifulSoup(driver.page_source, 'html.parser')
        section = soup_art.find('section', class_='module related-artworks related-masonry module-bottom-spacing')
        
        if section:
            artworks = section.find_all('li', class_='item')
            if len(artworks) > 7:
                more_artwork_url = artworks[7].find('a')['href']
                driver.get(more_artwork_url)
                wait.until(EC.presence_of_element_located((By.CLASS_NAME, 'results-grid')))
                soup_more_artwork = BeautifulSoup(driver.page_source, 'html.parser')
                section = soup_more_artwork.find('section', class_='results-grid scroll-load')
                
                if section:
                    all_artworks = section.find_all('li', class_='item')
                    for artwork in all_artworks:
                        title = artwork.find('span', class_='title').text.strip()
                        artwork_url = artwork.find('a')['href']
                        print(f'Title: {title}')
                        print(f'Link: {artwork_url}')
                        driver.get(artwork_url)
                        wait.until(EC.presence_of_element_located((By.CLASS_NAME, 'masonry_details')))
                        soup_artwork = BeautifulSoup(driver.page_source, 'html.parser')
                        masonry_details = soup_artwork.find('div', class_='masonry_details')
                        
                        if masonry_details:
                            details = {}
                            for item in masonry_details.find_all('div', class_='masonry-item'):
                                title = item.find('h5').text.strip()
                                value = item.find('p').text.strip()
                                details[title] = value
                            
                            for key, value in details.items():
                                print(f'{key}: {value}')
                        
                        driver.back()
                        time.sleep(1)
                    driver.back()
            else:
                for artwork in artworks:
                    title = artwork.find('span', class_='title').text.strip()
                    artwork_url = artwork.find('a')['href']
                    print(f'Title: {title}')
                    print(f'Link: {artwork_url}')
                    driver.get(artwork_url)
                    wait.until(EC.presence_of_element_located((By.CLASS_NAME, 'masonry_details')))
                    soup_artwork = BeautifulSoup(driver.page_source, 'html.parser')
                    masonry_details = soup_artwork.find('div', class_='masonry_details')
                    
                    if masonry_details:
                        details = {}
                        for item in masonry_details.find_all('div', class_='masonry-item'):
                            title = item.find('h5').text.strip()
                            value = item.find('p').text.strip()
                            details[title] = value
                        
                        for key, value in details.items():
                            print(f'{key}: {value}')
                    
                    driver.back()
        else:
            print("Artwork section not found.")
    
    scrap_artworks()
def main(url): 
    
    while True:
        
        driver.get(url)
        wait = WebDriverWait(driver, 2)
        soup = BeautifulSoup(driver.page_source,'html.parser')
       
        num_of_artist = soup.find('span',class_='num').text.strip()
        num_of_artist = num_of_artist.replace(',', '')
        number = int(num_of_artist)
        try:
            for j in range(1,number):                   
                artist_element = driver.find_element(By.XPATH, f'/html/body/main/section[2]/ul/li[{j}]/a')
                driver.execute_script("arguments[0].scrollIntoView();", artist_element)
                time.sleep(1)
                driver.execute_script("arguments[0].click();", artist_element)
                time.sleep(3)  
                WebDriverWait(driver, 5).until(EC.url_changes(url))
                print("__________________________________________________________________________________________")
                print(driver.current_url)
                
                scrap_artist(driver.current_url)
                driver.back()  
                time.sleep(2)  
        except:
            
            try:
                load_more_button = wait.until(EC.presence_of_element_located((By.CLASS_NAME, 'btn btn-default pagination-btn')))
                time.sleep(3)
                if load_more_button.is_displayed():
                      
                    artist_url=driver.current_url
                    
                    main(artist_url)
                else:
                    break
                    
            except:
            
                print("No more pages to load or error occurred:")
                pass
                



url = f'https://artuk.org/discover/artists/view_as/grid/page/1'        
main(url)
```
